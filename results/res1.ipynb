```python
import pandas as pd 
import numpy as np
import re 
from matplotlib import pyplot as plt 
```

```python
def get_data(file_list, privat_or_pub):

    place = []
    name = []
    score = []
    date = []

    if (privat_or_pub=='privat'):
        divider = 3
        for i in range(int(len(file_list) / divider)):
            place.append(file_list[divider*i+0])
            name.append(file_list[divider*i+1])
            score.append(file_list[divider*i+2])
    else:
        divider = 5
        for i in range(int(len(file_list) / divider)):
            place.append(file_list[divider*i+0])
            name.append(file_list[divider*i+1])
            score.append(file_list[divider*i+3])


    
    df = pd.DataFrame({
        'place_'+privat_or_pub: place,
        'name': name,
        'score_'+privat_or_pub: score,
    })
    df['place_'+privat_or_pub] = df['place_'+privat_or_pub].astype('int')
    df['score_'+privat_or_pub] = df['score_'+privat_or_pub].astype('float')

    return df
```

```python
file = open('./data/1_privat.txt', 'r') 
file = file.read() 
file_list = file.split('\n')
df_privat = get_data(file_list, 'privat')

file = open('./data/1_pub.txt', 'r') 
file = file.read() 
file_list = file.split('\n')
df_pub = get_data(file_list, 'pub')
```

```python
df = pd.merge(df_privat, df_pub, how='left', on='name')
```

```python
df['place_dif'] = df['place_privat'] - df['place_pub']
```
